```python
# Configurazione canvas
canvas = R.TCanvas("canvas", "Superimposed Test Statistics", 1200, 900)
canvas.SetMargin(0.15, 0.05, 0.12, 0.08)  # Margini: left, right, bottom, top
canvas.SetLogy()

# Disegno istogrammi
first_hist = True
i=0
for hist_name in attributes:
    if i>9:
        break
    hist = file.Get(hist_name.GetName())
    if not hist:
        continue
    
    if first_hist:
        # Impostazioni assi
        hist.SetTitle("Superimposed Test Statistics")
        hist.GetXaxis().SetTitle("TS")
        hist.GetYaxis().SetTitle("Probabilita'")
        
        # Formattazione matematica corretta
        hist.GetXaxis().SetTitleSize(0.045)
        hist.GetYaxis().SetTitleSize(0.045)
        hist.GetXaxis().SetLabelSize(0.04)
        hist.GetYaxis().SetLabelSize(0.04)
        
        # Posizionamento label
        hist.GetXaxis().SetTitleOffset(1.3)  # Aumentato ulteriormente
        hist.GetYaxis().SetTitleOffset(1.2)
        
        # Range assi
        hist.GetXaxis().SetRangeUser(0, 1.5)  # Ferma a 1.5 sull'asse X
        hist.GetYaxis().SetRangeUser(0, hist.GetMaximum()*1.2)
        
        hist.Draw("HIST")
        first_hist = False
    else:
        hist.Draw("HIST SAME")
        i = i+1

# Forza ridisegno
canvas.Modified()
canvas.Update()
```

```python
# Configurazione canvas
canvas = R.TCanvas("canvas", "Superimposed Test Statistics", 1200, 900)
canvas.SetMargin(0.15, 0.05, 0.12, 0.08)
canvas.SetLogy()  # Scala logaritmica asse Y

# Crea legenda
legend = R.TLegend(0.6, 0.5, 0.88, 0.88) # (x1, y1, x2, y2)
legend.SetBorderSize(0)
legend.SetFillStyle(0)
legend.SetTextSize(0.035)
legend.SetTextFont(42)

# Disegno istogrammi
first_hist = True
i = 0
for hist_name in attributes:
    hist = file.Get(hist_name.GetName())
    if not hist:
        continue
    if i>9: 
        break
    
    label =  hist_name.GetName()
    
    if first_hist:
        # Impostazioni assi
        hist.SetTitle("Superimposed Test Statistics")
        hist.GetXaxis().SetTitle("TS")
        hist.GetYaxis().SetTitle("Probabilit")
        
        hist.GetXaxis().SetTitleSize(0.045)
        hist.GetYaxis().SetTitleSize(0.045)
        hist.GetXaxis().SetLabelSize(0.04)
        hist.GetYaxis().SetLabelSize(0.04)
        hist.GetXaxis().SetTitleOffset(1.3)
        hist.GetYaxis().SetTitleOffset(1.2)
        
        hist.GetXaxis().SetRangeUser(0, 4)
#        hist.GetYaxis().SetRangeUser(1e-5, hist.GetMaximum() * 10)
        
        hist.Draw("HIST")
        legend.AddEntry(hist, label, "l")
        first_hist = False
    else:
        hist.Draw("HIST SAME")
        legend.AddEntry(hist, label, "l")
        i+=1

# Disegna la legenda
legend.Draw()

# Forza ridisegno
canvas.Modified()
canvas.Update()
canvas.SaveAs("presentation TS.png")
```

```python
#Chiudi il canvas esistente se gi presente
if 'canvas' in globals():
    canvas.Close()

# Crea un nuovo canvas
canvas = R.TCanvas("canvas", "Istogrammi sovrapposti", 800, 600)
R.gStyle.SetOptStat(0)  # Disabilita il box delle statistiche

# Imposta i colori per i diversi istogrammi
colors = [R.kRed, R.kBlue, R.kGreen, R.kMagenta, R.kBlack]

# Disegna ogni istogramma con "SAME"
first_hist = True
for i, hist_name in enumerate(attributes):
    hist = file.Get(hist_name.GetName())
    if not hist:
        print(f"Attenzione: {hist_name.GetName()} non trovato nel file!")
        continue
    
    hist.SetLineColor(colors[i % len(colors)])  # Assegna un colore diverso
    hist.SetLineWidth(2)  # Spessore linea
    
    if first_hist:
        hist.Draw()  # Disegna il primo istogramma
        first_hist = False
    else:
        
        hist.Draw("SAME")  # Sovrapponi gli altri

# Imposta i range degli assi (opzionale)
if not first_hist:  # Se almeno un istogramma  stato disegnato
    canvas.GetPad(0).SetGridx()
    canvas.GetPad(0).SetGridy()
    canvas.Modified()
    canvas.Update()

# Aggiungi una legenda (opzionale)
leg = R.TLegend(0.7, 0.7, 0.9, 0.9)
for i, hist_name in enumerate(attributes):
    hist = file.Get(hist_name.GetName())
    if hist:
        leg.AddEntry(hist, hist_name.GetName(), "l")
leg.Draw()

# Mantieni la finestra aperta
canvas.Draw()
R.gApplication.Run()
```
