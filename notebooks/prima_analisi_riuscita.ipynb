```python
if 'canvas' in globals():
    canvas.Close()

canvas = R.TCanvas('c1', 'Dynamic Filling Example', 400, 30, 1600, 800)
canvas.Divide(2, 1)

R.gStyle.SetOptStat(0)

# --- Primo plot ---
canvas.cd(1)
pad1 = canvas.GetPad(1)
pad1.SetLeftMargin(0.2)
pad1.SetRightMargin(0.15)
pad1.SetBottomMargin(0.18)
pad1.SetTopMargin(0.1)

h_2d.SetTitle("Neutrino distribution; sin(#delta) ; Energy [GeV]")
h_2d.GetXaxis().SetTitleOffset(1.5)
h_2d.GetYaxis().SetTitleOffset(1.5)
h_2d.Draw("COLZ")

# --- Secondo plot ---
canvas.cd(2)
pad2 = canvas.GetPad(2)
pad2.SetLeftMargin(0.2)
pad2.SetRightMargin(0.15)
pad2.SetBottomMargin(0.18)
pad2.SetTopMargin(0.1)

h_2d1.SetTitle("Second distribution; sin(#delta) ; Energy [GeV]")
h_2d1.GetXaxis().SetTitleOffset(1.5)
h_2d1.GetYaxis().SetTitleOffset(1.5)
h_2d1.Draw("COLZ")

canvas.Update()
canvas.Show()
```

```python
if 'canvas' in globals():
    canvas.Close()

canvas = R.TCanvas('c1', 'Single Plot Example', 400, 30, 1200, 800)

R.gStyle.SetOptStat(0)

canvas.SetLeftMargin(0.2)
canvas.SetRightMargin(0.15)
canvas.SetBottomMargin(0.18)
canvas.SetTopMargin(0.1)

h_2d.SetTitle("signal Distribution in KM3NeT ")

h_2d.SetYTitle("log_{10}(Energy [GeV])")
h_2d.SetXTitle("#alpha")

h_2d.GetXaxis().SetTitleOffset(1.5)
h_2d.GetYaxis().SetTitleOffset(1.5)
h_2d.Draw("COLZ")

canvas.Update()
canvas.Show()
canvas.SaveAs("An example of a signal (fixed source) .png")
```
