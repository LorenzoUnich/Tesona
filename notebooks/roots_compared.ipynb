```python
import ROOT as R
import matplotlib.pyplot as plt
import os 

def find_root_files(path):
	files = os.listdir(path)
	root_files = []
	print("you have the following files to analyse")
	for i, f in enumerate(files):

		if f[-4:] == "root":
			print("file n. ", i , " file" , f)
			root_files.append(f)

	root_files.sort()
	return root_files

path = "/home/lori/Desktop/Tesi/root_declination_no_bands"


root_files_no_bands = find_root_files(path)

```

```python
def get_2d_hist(path,names):
    hists = []
    for name in names:
        
        file = R.TFile.Open(path+"/"+name)
        attributes = file.GetListOfKeys()
        for attribute in attributes:
            obj = file.Get(attribute.GetName())
            print(attribute.GetName())
            if obj.InheritsFrom("TH2"):
                clone = obj.Clone(f"{obj.GetName()}_clone")
                clone.SetDirectory(0)
                hists.append(clone)
        file.Close()  # ora puoi chiudere
    return hists

path1 = "/home/lori/Desktop/Tesi/root_declination_bands"
root_files_bands = find_root_files(path1)

```

```python
hist_bands= get_2d_hist(path1, root_files_bands)
hist_no_bands = get_2d_hist(path, root_files_no_bands)
```

```python
hist_bands
```

```python

only_background_nb = []
only_background_b = []
for hist in hist_bands:
    name = hist.GetName()
    print(name)
    if name[13] == "b":
        only_background_b.append(hist)
for hist in hist_no_bands:
    name = hist.GetName()
    if name[13] == "b":
        only_background_nb.append(hist)
```

```python
only_signal_b = []
only_signal_nb = []


for hist in hist_bands:
    name = hist.GetName()
    print(name)
    if name[13] == "s":
        only_signal_b.append(hist)
for hist in hist_no_bands:
    name = hist.GetName()
    if name[13] == "s":
        only_signal_nb.append(hist)
```

```python
if "canvas" in globals():
    canvas.Close()
canvas = R.TCanvas("firstplot", '', 1800, 800)
for i, h_b in enumerate(only_background_b):
    h_nb= only_background_nb[i]
    R.gStyle.SetOptStat(0)

    # Proiezioni
    proj_bands = h_b.ProjectionY("Bands")
    proj_no_bands = h_nb.ProjectionY("Nobands")
    proj_bands.SetTitle(f"Background for source {i}")
    

    # Colori
    proj_bands.SetLineColor(R.kRed)
    proj_no_bands.SetLineColor(R.kBlue)
    print("source ", i)
    print("band integral",  proj_bands.Integral())
    print("not band integral",  proj_no_bands.Integral())

    #proj2.SetXLabel("piru")
    # Disegno
    proj_bands.Draw()
    proj_no_bands.Draw("same")

    # Legenda manuale
    legend = R.TLegend(0.7, 0.7, 0.9, 0.9)  # posizione: (x1, y1, x2, y2)
    legend.AddEntry(proj_bands, "With bands", "l")  # "l" = line
    legend.AddEntry(proj_no_bands, "Without bands", "l")
    legend.Draw()
    canvas.Show()
    canvas.SaveAs(f"compare_histograms/bkgnd/{i} comparison background.png")
```

```python
if "canvas" in globals():
    canvas.Close()
canvas = R.TCanvas("firstplot", '', 1800, 800)
for i, h_b in enumerate(only_signal_b):
    h_nb= only_signal_nb[i]
    R.gStyle.SetOptStat(0)

    # Proiezioni
    proj_bands = h_b.ProjectionY("Bands")
    proj_no_bands = h_nb.ProjectionY("Nobands")
    proj_bands.SetTitle(f"Signal for source {i}")
    

    # Colori
    proj_bands.SetLineColor(R.kRed)
    proj_no_bands.SetLineColor(R.kBlue)
    print("source ", i)
    print("band integral",  proj_bands.Integral())
    print("not band integral",  proj_no_bands.Integral())

    #proj2.SetXLabel("piru")
    # Disegno
    proj_bands.Draw()
    proj_no_bands.Draw("same")

    # Legenda manuale
    legend = R.TLegend(0.7, 0.7, 0.9, 0.9)  # posizione: (x1, y1, x2, y2)
    legend.AddEntry(proj_bands, "With bands", "l")  # "l" = line

    legend.AddEntry(proj_no_bands, "Without bands", "l")
    legend.Draw()
    canvas.Show()
    canvas.SaveAs(f"compare_histograms/sgn/{i} comparison signal.png")
```

```python
if "canvas" in globals():
    canvas.Close()
canvas = R.TCanvas("firstplot", '', 1800, 800)
for i, h_b in enumerate(only_background_b):
    h_nb= only_background_nb[i]
    sgn = only_signal_b[i]
    R.gStyle.SetOptStat(0)

    # Proiezioni
    proj_bands = h_b.ProjectionY("Bands")
    proj_no_bands = h_nb.ProjectionY("Nobands")
    proj_signal = sgn.ProjectionY("Signal")

    proj_bands.SetTitle(f"Background for source {i}")
    

    # Colori
    proj_bands.SetLineColor(R.kRed)
    proj_no_bands.SetLineColor(R.kBlue)
    proj_signal.SetLineColor(R.kViolet)
    print("source ", i)
    print("band integral",  proj_bands.Integral())
    print("not band integral",  proj_no_bands.Integral())

    #proj2.SetXLabel("piru")
    # Disegno
    proj_bands.Draw()
    proj_no_bands.Draw("same")
    proj_signal.Draw("same")
    proj_bands.SetLineWidth(3)
    proj_no_bands.SetLineWidth(3)
    proj_signal.SetLineWidth(6)
    proj_signal.SetLineStyle(2)
    # Legenda manuale
    legend = R.TLegend(0.7, 0.7, 0.9, 0.9)  # posizione: (x1, y1, x2, y2)
    legend.AddEntry(proj_bands, "With bands", "l")  # "l" = line
    legend.AddEntry(proj_no_bands, "Without bands", "l")
    legend.AddEntry(proj_signal, "Signal", "l")
    
    legend.Draw()
    canvas.Show()
    canvas.SaveAs(f"compare_histograms/sgn_bkgnd/{i} comparison background.png")
```
