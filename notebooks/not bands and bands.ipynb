```python
import matplotlib.pyplot as plt  
import pandas as pd
import numpy as np

file =  "extracted_summary.txt"
file_bands = "extracted_summary_bands.txt"

df = pd.read_csv(file,  sep="\t")
df_bands = pd.read_csv(file_bands,  sep="\t")
```

```python
df_bands.keys()
```

```python
plt.figure(figsize =(5,3.5), dpi = 170)
plt.scatter(df_bands.source_declination,  df["ss90"],label = "not in bands")
plt.scatter(df_bands.source_declination, df_bands.ss90,  label= "in bands")
plt.legend()
plt.xlabel("Source declination")
plt.ylabel("ss level")
```

```python
plt.figure(figsize =(5,3.5), dpi = 170)
plt.scatter(df_bands.source_declination , df["discovery"],label = "not in bands")
plt.scatter(df_bands.source_declination, df_bands.discovery,  label= "in bands")
plt.legend()
plt.xlabel("Source declination")
plt.ylabel("Discovery")
```

```python
discovery_fraction =   [ df_bands.discovery[i]/df.discovery[i] for i in range(len(df_bands.discovery))]
```

```python
plt.figure(figsize =(5,3.5), dpi = 170)
plt.scatter(df_bands.source_declination , discovery_fraction)
plt.xlabel("Discovery fraction (bands/no bands)")
```

```python
plt.figure(figsize =(5,3.5), dpi = 170)
plt.scatter(df["i"],  df["discovery"],label = "not in bands")
plt.scatter(df_bands.i, df_bands.discovery,  label= "in bands")
plt.legend()
plt.xlabel("Source number")
plt.ylabel("Discovery")
```
