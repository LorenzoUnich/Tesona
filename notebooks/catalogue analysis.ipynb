```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("catalogo_pulito.csv",sep=",")
df
```

```python
declins= []
disc = []
taken = 0
for i in df["δ(Jy)"][1:]:
    try:
        taken+=1
        f = float(i)
        declins.append(f)
    except:
        disc.append(i)
        continue
print(disc, taken)
```

```python
plt.hist(declins, bins= 100)
plt.xlabel(r"Declination ($\delta$)")
plt.yscale("log")
```
