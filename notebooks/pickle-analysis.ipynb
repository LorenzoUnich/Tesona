```python
def gaussian(x, pars):
    x = x[0]
    mu = pars[0]
    sigma = pars[1]
    N = pars[2]

    if sigma ==0:
        return 0 
    if N <=0:
        return 0
    return float(N*np.exp((-((x-mu)/(sigma))**2)/2))/(np.sqrt(2*np.pi)*sigma)

def double_gaussian(x, pars):
    if isinstance(x, (list, tuple, np.ndarray)):
        x = x[0]
    mu_1 = pars[0]
    sigma_1 = pars[1]
    mu_2 = pars[2]
    sigma_2 = pars[3]
    N1 = pars[4]
    N2 = pars[5]


    if sigma_1 <= 0 or sigma_2 <= 0:
        return 0
    if N1 <=0 or N2<=0:
        return 0
    return float(gaussian(x, (mu_1, sigma_1, N1))+ gaussian(x, (mu_2, sigma_2,N2)))
def triple_gauss(x, pars):
    if isinstance(x, (list, tuple, np.ndarray)):
        x = x[0]
    mu_1 = pars[0]
    sigma_1 = pars[1]
    mu_2 = pars[2]
    sigma_2 = pars[3]
    N1 = pars[4]
    N2 = pars[5]
    mu_3 = pars[6]
    sigma_3 = pars[7]
    N3 = pars[8]
    if sigma_1 <= 0 or sigma_2 <= 0:
        return 0
    if N1 <=0 or N2<=0:
        return 0
    if N3<=0 or sigma_3<=0:
        return 0
    return float(gaussian(x, (mu_1, sigma_1, N1))+ gaussian(x, (mu_2, sigma_2,N2)))
```

```python
list_of_bands = ["-1° to -0.6°","-0.6° to -0.2°","-0.2° to 0.2°", "0.2° to 0.4"]
```

```python
canvas = R.TCanvas('c1', 'Dynamic Filling Example', 400, 30, 1600, 800)
canvas.Divide(2, 2)
R.gStyle.SetOptStat(0)

histos = []  # Per mantenere gli oggetti in memoria
files = []   # idem con patate

for i, fname in enumerate(root_files):
    f = R.TFile.Open(fname)
    files.append(f)  # tieni aperto il file

    keys = f.GetListOfKeys()
    if not keys or keys.GetSize() == 0:
        print(f" Nessun oggetto trovato in {fname}")
        continue

    thing_I_need = keys[0].GetName()
    h = f.Get(thing_I_need)

    if not h or not h.InheritsFrom("TH1"):
        print(f" Oggetto {thing_I_need} non è un istogramma valido in {fname}")
        continue

    # Clonalo e staccalo dal file
    h_clone = h.Clone(f"{thing_I_need}_clone_{i}")
    h_clone.SetDirectory(0)
    h_clone.SetTitle("declination from " +list_of_bands[i])
    histos.append(h_clone)

    canvas.cd(i + 1)
    h_clone.Draw("COLZ")

canvas.Update()
canvas.SaveAs("all_histograms_multi_.png")
```
